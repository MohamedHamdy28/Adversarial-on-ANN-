```python
import os
import thinkdsp as tp
import numpy as np
import librosa
from librosa.display import specshow
import matplotlib.pyplot as plt
```

```python
FILE_PATH_INPUT = 'train_audio/yes'
FILE_PATH_OUTPUT = 'train_spec/yes'
FILE_PATH_INPUT_LIST = os.listdir(FILE_PATH_INPUT)
FILE_PATH_INPUT_LIST[:10]
```

```python
for i in range(len(FILE_PATH_INPUT_LIST)):
    wave = tp.read_wave(f"{FILE_PATH_INPUT}/{FILE_PATH_INPUT_LIST[i]}")
    spectrogram = wave.make_spectrogram(seg_length=1024)
    spectrogram.plot(high=5000)
    plt.axis('off')
    plt.savefig(f'{FILE_PATH_OUTPUT}/{i}.png', bbox_inches='tight', pad_inches=0)
```

# Generating spectograms

```python
FOLDER_PATH_INPUT = 'my_audio'
FOLDER_PATH_OUTPUT = 'my_specto'

for folder in os.listdir(FOLDER_PATH_INPUT):
    if not os.path.exists(rf'{FOLDER_PATH_OUTPUT}/{folder}'):
        os.makedirs(rf'{FOLDER_PATH_OUTPUT}/{folder}')
    for file in os.listdir(rf'{FOLDER_PATH_INPUT}/{folder}'):
        wave = tp.read_wave(f"{FOLDER_PATH_INPUT}/{folder}/{file}")
        spectrogram = wave.make_spectrogram(seg_length=1024)
        spectrogram.plot(high=5000)
        plt.axis('off')
        plt.savefig(f'{FOLDER_PATH_OUTPUT}/{folder}/{file[:-4]}.png', bbox_inches='tight', pad_inches=0)
```
